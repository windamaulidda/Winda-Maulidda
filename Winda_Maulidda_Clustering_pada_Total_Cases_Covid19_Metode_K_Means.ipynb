# Clusterisasi Dataset Covid 19 di Indonesia dengan metode K-Means

Menginputkan package dan input data yang dimiliki

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Membaca file 
covd19_df = pd.read_excel('covid_19_indonesia_time_series_all.xlsx')
# Keterangan isi index 
print(covd19_df.keys())
covd19_df
```

Hanya mengambil beberapa data yaitu Date, Location, Island, New Deaths, dan Total Cases

```python
# Menyajikan Tanggal,Lokasi,Kasus Baru,Kasus Kematian,Total kasus,Total kematian dan Island
ind_cov_df = pd.DataFrame(covd19_df[['Date','Location', 'Island','New Cases', 'New Deaths','Total Cases', 'Total Deaths']])
ind_cov_df["Date"] = ind_cov_df["Date"].astype('datetime64[ns]')
ind_cov_df.sort_values(by=["Date"])
```

Hanya mengambil data Total Cases dan Total Kematian per-provinsi dalam kurun waktu Maret sampai Desember

```python
# Total kasus dan Total kematian per-provinsi dalam kurun waktu MARET s/d DESEMBER.
ind_cov19_df = pd.DataFrame(covd19_df[['Date','Location','Island','Total Cases','Total Deaths']])
ind_cov19_df["Date"] = ind_cov19_df["Date"].astype('datetime64[ns]')
ind_cov19_df.sort_values(by=["Date"])
```

Memanggil file csv baru yang berisi Total Cases per-provinsi

```python
Cases = pd.read_csv("Total_Cases.csv")
Cases
```

Mengubah variabel yang sebelumnya berbentuk data frame menjadi array

```python
a= Cases.drop(Cases.columns[0], axis=1)
b= np.array(a)
print(b)
```

Menentukan jumlah klasternya, Misal menjadi 3 klaster

```python
#Menentukan dan mengkonfigurasi fungsi kmeans
kmeans =KMeans(n_clusters = 3, random_state=123)
#Menentukan kluster dari data
kmeans.fit(b)
```

Mencari nilai pusat dari masing-masing klaster

```python
print(kmeans.cluster_centers_)
```

Kemudian tampilkan hasil klaster dan tambahkan kolom klaster kedalam data frame

```python
#Menampilkan hasil kluster
print(kmeans.labels_)
#Menambahkan kolom "klaster" dalam data frame
Cases["klaster"] = kmeans.labels_
print(Cases)
```
